## Reading

```python
# Import the needed libraries
import cv2
import matplotlib.pyplot as plt
```

```python
def show_image(left,right):
  plt.figure(figsize=(10,5))

  plt.subplot(1,2,1)
  plt.imshow(left)
  plt.title('Left Image')

  plt.subplot(1,2,2)
  plt.imshow(right)
  plt.title('Right Image')

  plt.show()
```

```python
# read the images and show them in one figure in RGB
left=cv2.imread('stitch_left.jpg')
right=cv2.imread('stitch_right.jpg')

left=cv2.cvtColor(left,cv2.COLOR_BGR2RGB)
right=cv2.cvtColor(right,cv2.COLOR_BGR2RGB)

show_image(left,right)
```

## Image Stitching

```python
# define and use the stitcher
stitcher=cv2.Stitcher_create()
status, result=stitcher.stitch([left,right])
# print the status of the operation
print(status)
```

```python
# Show the stitiched (result) image
plt.imshow(result)
plt.title('Stitched Result')
plt.show()
```

```python
def stitcher_fun(scale):
  right_new_scale=cv2.resize(right,None,fx=scale,fy=scale)
  stitcher=cv2.Stitcher_create()
  status, result=stitcher.stitch([left,right_new_scale])
  plt.imshow(result)
  plt.title('Stitched Result')
  plt.show()
```

```python
# use the scaling function, scale the right image by a factor of 0.5 and show the results of stitching it with the unchanged left image (if there is a result)
stitcher_fun(0.5)
```

```python
# use the scaling function, scale the right image by a factor of 0.75 and show the results of stitching it with the unchanged left image (if there is a result)
stitcher_fun(0.75)
```

## Depth

```python
# Load the stereo images and show them in RGB
stereo_left=cv2.imread('stereo_left.png')
stereo_right=cv2.imread('stereo_right.png')

stereo_left=cv2.cvtColor(stereo_left,cv2.COLOR_BGR2RGB)
stereo_right=cv2.cvtColor(stereo_right,cv2.COLOR_BGR2RGB)

show_image(stereo_left,stereo_right)
```

```python
# convert both images to a suitable colour space
stereo_left=cv2.cvtColor(stereo_left,cv2.COLOR_BGR2GRAY)
stereo_right=cv2.cvtColor(stereo_right,cv2.COLOR_BGR2GRAY)

# Define and use the stereo function
stereo=cv2.StereoBM_create(numDisparities=112,blockSize=19)
disparity=stereo.compute(stereo_left,stereo_right)

# Using a focal length of 13000 and a baseline of 120mm, find the depth in meters
disparity[disparity==0]=0.1
focal_lenght=13000
baseline=0.12
depth=focal_lenght*baseline/disparity

# show the depth map
plt.imshow(depth)
plt.title('The depth map')
plt.show()
```
