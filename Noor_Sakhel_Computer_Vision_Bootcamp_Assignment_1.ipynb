# Computer Vision Bootcamp: Assignment 1

## Reading

Take a photo of any object in your room using your mobile or laptop's camera

Upload the image into Colab. Then read the image

```python
# Import the needed libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# read the image
from google.colab import files
upload=files.upload()
uploaded_image=list(upload.keys())[0]

I =cv2.imread(uploaded_image)
```

Resize the image to 512*512 pixels using the resize function in OpenCV

```python
resized_image = cv2.resize(I,(512,512))
```

## Converting

```python
# Convert the reseized image to RGB
rgb_image = cv2.cvtColor(resized_image,cv2.COLOR_BGR2RGB)

# show the rgb image
plt.imshow(rgb_image)
plt.show()
```

```python
# Convert the reseized image to grayscale
gray_image =cv2.cvtColor(resized_image,cv2.COLOR_BGR2GRAY)

# show the reseized gray image
plt.imshow(gray_image,cmap='gray')
plt.show()
```

```python
# Split the reseized image into its three color channels
B, G, R=cv2.split(resized_image)

# show the channels in one plot
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(R,cmap='Reds_r')

plt.subplot(1,3,2)
plt.imshow(G,cmap='Greens_r')

plt.subplot(1,3,3)
plt.imshow(B, cmap='Blues_r')
plt.show()
```

## Convolutions

### Smoothing

```python
# convole the image with Box mask of size 3, 9, 17, 25 and plot the four results in one plot

mask=np.ones((3,3))/9
one_image=cv2.filter2D(rgb_image,-1,mask)

mask=np.ones((9,9))/81
two_image=cv2.filter2D(rgb_image,-1,mask)

mask=np.ones((17,17))/17**2
three_image=cv2.filter2D(rgb_image,-1,mask)

mask=np.ones((25,25))/25**2
four_image=cv2.filter2D(rgb_image,-1,mask)

plt.figure(figsize=(20,4))

plt.subplot(1,4,1)
plt.imshow(one_image)

plt.subplot(1,4,2)
plt.imshow(two_image)

plt.subplot(1,4,3)
plt.imshow(three_image)

plt.subplot(1,4,4)
plt.imshow(four_image)

plt.show()
```

```python
# convole the image with Gaussian mask of Scale 3, 9, 17, 25 (choose suitable mask size) and plot the four results in one plot
scale=3
size=19
one_gaussian_image=cv2.GaussianBlur(rgb_image,(size,size),scale)

scale=9
size=55
two_gaussian_image=cv2.GaussianBlur(rgb_image,(size,size),scale)

scale=17
size=103
three_gaussian_image=cv2.GaussianBlur(rgb_image,(size,size),scale)

scale=25
size=151
four_gaussian_image=cv2.GaussianBlur(rgb_image,(size,size),scale)

plt.figure(figsize=(20,4))

plt.subplot(1,4,1)
plt.imshow(one_gaussian_image)

plt.subplot(1,4,2)
plt.imshow(two_gaussian_image)

plt.subplot(1,4,3)
plt.imshow(three_gaussian_image)

plt.subplot(1,4,4)
plt.imshow(four_gaussian_image)

plt.show()
```

### Difference

```python
# convole the image with horizontal mask to catch vertical edges
mask_v=np.array([[-1,0,1],
                 [-2,0,2],
                 [-1,0,1]])
edges_v=cv2.filter2D(gray_image,-1,mask_v)

# convole the image with vertical mask to catch horizontal edges
mask_h=np.transpose(mask_v)
edges_h=cv2.filter2D(gray_image,-1,mask_h)

# convole the image with a Laplacian mask to catch all edges
mask_lap=np.array([[-1,-1,-1],
                   [-1, 8,-1],
                   [-1,-1,-1]])
edges_lap=cv2.filter2D(gray_image,-1,mask_lap)

# plot the three figures along with the gray scale resized image in one plot (4 images in total: grayscale, vertical edges, horizontal edges, and all edges)
plt.figure(figsize=(20,4))
plt.subplot(1,4,1)
plt.imshow(gray_image,cmap='gray')

plt.subplot(1,4,2)
plt.imshow(edges_v,cmap='gray')

plt.subplot(1,4,3)
plt.imshow(edges_h,cmap='gray')

plt.subplot(1,4,4)
plt.imshow(edges_lap,cmap='gray')

plt.show()
```
